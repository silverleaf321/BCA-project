# Loading Data

```python
url = 'https://media.githubusercontent.com/media/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn/master/Chapter01/bike_rentals.csv'
import pandas as pd # import pandas for library building
df = pd.read_csv(url) # load data
df = df.fillna(df.median()) #clear null values
df.head() # show data
```

## Split Data into X and y

```python
X = df.iloc[:, 2:-3]
y = df.iloc[:,-1:]

```

# Machine Learning

```python
X = df.iloc[:, 2 : -3]
y = df.iloc[:, -1 :]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
model = XGBRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
print(mse)
print(rmse)
```

```python
model = XGBRegressor()
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_train, y_train, scoring = "neg_mean_squared_error")
rmse_scores = (-scores) ** 0.5
print(rmse_scores)
rmse_mean = rmse_scores.mean()
print(rmse_mean)
```

```python
from sklearn.model_selection import GridSearchCV
model = XGBRegressor()

# establish a dictionary of parameters to check
params = {"max_depth": [2, 3, 4, 5, 6, 8, 10]}
# cross validation
grid_reg = GridSearchCV(model, params, scoring = "neg_mean_squared_error", cv = 5)

# fit model on data
grid_reg.fit(X_train, y_train)
```

```python
def grid_search(model, params):
  # show best parameters
  best_params_ = grid_reg.best_params_

  # get best scores
  scores = grid_reg.best_score_
  rmse = (-scores) ** 0.5
  best_score = rmse

  print(best_params_)
  print(best_score)
```

```python
model = XGBRegressor()
params = {"max_depth": [2, 3, 4, 5, 6, 8, 10, 12]}

grid_search(model, params)
```

```python
model = XGBRegressor(max_depth = 2)
params = {"subsample": [0.2, 0.4, 0.6, 0.8, 1]}

grid_search(model, params)
```

```python
model = XGBRegressor(max_depth = 2)
params = {"subsample": [0.2, 0.4, 0.6, 0.8, 1],
          "max_depth": [2, 3, 4, 5, 6, 8, 10, 12]}

grid_search(model, params)
```
