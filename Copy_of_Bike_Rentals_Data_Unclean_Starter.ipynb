# Loading Data

```python
url = 'https://media.githubusercontent.com/media/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn/master/Chapter01/bike_rentals.csv'
import pandas as pd
df = pd.read_csv(url)
df.head()
```

# Cleaning the Data

```python
df.info()
```

```python
df = df.fillna('median')
df.info()
```

`df.info()` indicates that all null values have been cleared.

# Split Data into X and y

```python
# set y as the last column
y = df.iloc[:,-1:]
y.head()
```

```python
# Set X: start with 2nd column include all but last 3
X = df.iloc[:, 2:-3]
X.head()
```

# Verify Data is ready for ML

```python
X.info()
```

All input data, the X columns, are numeric with no null values.

```python
y.info()
```

All output data, the y, is numeric with no null values.
