```python
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)

X_train.head()
```

```python
X_train.shape
```

```python
X_train.shape[1]
```

```python
# import general model and style of layer
from keras.models import Sequential
from keras.layers import Dense

# initialize model
model = Sequential()

# get number of cols (needed for input)
num_cols = X_train.shape[1]

# add a hidden layer with 16 nodes
model.add(Dense(16, input_shape=(num_cols,), activation='relu')) # specify num of cols as input_shape for 1st layer only

# add a hidden layer with 4 nodes
model.add(Dense(4, activation='relu'))

# add output layer (always 1 node)
model.add(Dense(1))

print(model.summary())
```

```python
model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, epochs=40)

mse = model.evaluate(X_test, y_test)
print(mse**0.5)
```
